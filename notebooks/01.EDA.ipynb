### Lexical Analysis

We explore the most frequent words in the responses to identify common lexical patterns and differences between safe and unsafe content.

#### Tokenization and Text Preparation

```python
import re
from collections import Counter

def simple_tokenize(text):
    text = str(text).lower()
    text = re.sub(r"[^\w\s]", " ", text)
    return text.split()
```

#### Most Frequent Words in Prompts

```python
all_words = " ".join(train_df["prompt"]).lower().split()
word_freq = Counter(all_words)

word_freq.most_common(20)
```

```python
common_words = pd.DataFrame(word_freq.most_common(20), columns=["word","count"])

sns.barplot(data=common_words, x="count", y="word")
plt.title("Top 20 Most Frequent Words in Prompts")
plt.xlabel("Count")
plt.ylabel("Word")
plt.show()
```

#### Most Frequent Words in Responses

```python
response_words = []
for text in train_df["response"]:
    response_words.extend(simple_tokenize(text))

response_freq = Counter(response_words)
response_common = pd.DataFrame(response_freq.most_common(20), columns=["word", "count"])

plt.figure(figsize=(8,5))
sns.barplot(data=response_common, x="count", y="word")
plt.title("Top 20 most frequent words in responses")
plt.tight_layout()
plt.show()
```

#### Most Frequent Words by Safety Class

```python
safe_words = []
unsafe_words = []

for text in train_df[train_df["is_safe"] == True]["response"]:
    safe_words.extend(simple_tokenize(text))

for text in train_df[train_df["is_safe"] == False]["response"]:
    unsafe_words.extend(simple_tokenize(text))

safe_common = pd.DataFrame(Counter(safe_words).most_common(15), columns=["word", "count"])
unsafe_common = pd.DataFrame(Counter(unsafe_words).most_common(15), columns=["word", "count"])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.set_style("white")

sns.barplot(
    data=safe_common,
    x="count",
    y="word",
    ax=axes[0],
    color="#4CAF50"
)
axes[0].set_title("Top Words in Safe Responses", fontsize=12, weight="bold")
axes[0].set_xlabel("Count")
axes[0].set_ylabel("Word")

sns.barplot(
    data=unsafe_common,
    x="count",
    y="word",
    ax=axes[1],
    color="#E53935"
)
axes[1].set_title("Top Words in Unsafe Responses", fontsize=12, weight="bold")
axes[1].set_xlabel("Count")
axes[1].set_ylabel("Word")

sns.despine()
plt.tight_layout()
plt.show()
```

```python
from spacy.lang.es.stop_words import STOP_WORDS as ES_STOPWORDS
from spacy.lang.ca.stop_words import STOP_WORDS as CA_STOPWORDS

STOPWORDS = set(ES_STOPWORDS) | set(CA_STOPWORDS)
STOPWORDS |= {"d", "l", "n", "q", "x", "t", "m", "u", "p", "si", "sí"}

def tokenize_text(text):
    if pd.isna(text):
        return []
    text = text.lower()
    tokens = re.findall(r"\b[a-zà-úüïç·]{2,}\b", text)
    return [tok for tok in tokens if tok not in STOPWORDS and len(tok) > 2]

def get_counter(texts):
    counter = Counter()
    total = 0
    for text in texts:
        toks = tokenize_text(text)
        counter.update(toks)
        total += len(toks)
    return counter, total

safe_texts = train_df.loc[train_df["is_safe"] == 1, "response"]
unsafe_texts = train_df.loc[train_df["is_safe"] == 0, "response"]

safe_counter, safe_total = get_counter(safe_texts)
unsafe_counter, unsafe_total = get_counter(unsafe_texts)

vocab = set(safe_counter.keys()) | set(unsafe_counter.keys())

rows = []
for word in vocab:
    safe_freq = safe_counter[word] / safe_total if safe_total > 0 else 0
    unsafe_freq = unsafe_counter[word] / unsafe_total if unsafe_total > 0 else 0
    diff = unsafe_freq - safe_freq
    rows.append((word, safe_freq, unsafe_freq, diff))

diff_df = pd.DataFrame(rows, columns=["word", "safe_freq", "unsafe_freq", "diff"])

top_unsafe = diff_df.sort_values("diff", ascending=False).head(12).copy()

top_safe = diff_df.sort_values("diff", ascending=True).head(12).copy()
top_safe["diff"] = top_safe["diff"].abs()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(data=top_safe, x="diff", y="word", ax=axes[0], color="#5AA05A")
axes[0].set_title("Most Distinctive Words in Safe Responses")
axes[0].set_xlabel("Frequency difference")
axes[0].set_ylabel("Word")

sns.barplot(data=top_unsafe, x="diff", y="word", ax=axes[1], color="#C94A44")
axes[1].set_title("Most Distinctive Words in Unsafe Responses")
axes[1].set_xlabel("Frequency difference")
axes[1].set_ylabel("Word")

sns.despine()
plt.tight_layout()
plt.show()
```

These observations support the use of TF-IDF features and linear classifiers (e.g., Logistic Regression or SVM), which can better capture discriminative lexical patterns.

- The most frequent words are largely stopwords, which is expected.
- There are no immediately obvious discriminative words between safe and unsafe responses at this level.
- This suggests that simple frequency-based features may not be sufficient, and more advanced representations (TF-IDF) are needed.

### TF-IDF exploration

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000)

X_tfidf = vectorizer.fit_transform(train_df["prompt"])

feature_names = vectorizer.get_feature_names_out()
```

```python
X_tfidf.shape
```

```python
import numpy as np

mean_tfidf = np.asarray(X_tfidf.mean(axis=0)).flatten()
top_indices = mean_tfidf.argsort()[-20:]

[top_indices]
```

```python
top_terms = pd.DataFrame({
    "term": feature_names[top_indices],
    "score": mean_tfidf[top_indices]
}).sort_values("score", ascending=False)

top_terms
```
